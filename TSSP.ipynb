```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import yfinance as yf
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_squared_error
```

```python
# Ask user for input (e.g. "AAPL", "TSLA", "RELIANCE.NS")
ticker = input("Enter stock ticker (e.g., AAPL, TSLA, INFY.NS): ")

# Download last 5 years of data until today
data = yf.download(ticker, start='2018-01-01', end=None)  # None = today
data.head()
```

```python
import datetime

# Automatically use today's date
today = datetime.date.today()
data = yf.download(ticker, start='2018-01-01', end=today)
```
